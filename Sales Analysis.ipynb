<span style="color: RED; font-size: 60px; font-weight: bold;">EXPLORATORY DATA ANALYSIS</span>

<span style="color: BLUE; font-size: 30px; font-weight: bold;">GENDER</span>

```python
# PLOTTING BAR CHART FOR GENDER AND ITS COUNT
ax = sns.countplot(x='Gender',data = df,palette={'M':'blue','F':'orange'},hue='Gender')

for bars in ax.containers:
    ax.bar_label(bars)
```

```python
# PLOTTING BAR GRAPH FOR GENDER VS TOTAL AMOUNT
sales_gen= df.groupby(['Gender'],as_index=False,)['Amount'].sum().sort_values(by='Amount',ascending=False)
sns.barplot(x='Gender',y='Amount',data=sales_gen,palette={'M':'red','F':'skyblue'},hue='Gender')
```

<span style="color: blue; font-size: 30px; font-weight: bold;">MARITAL STATUS</span>

```python
ax= sns.countplot(data = df, x = 'Age Group',hue = 'Gender',palette='coolwarm')    

for bars in ax.containers:
    ax.bar_label(bars)

```

```python
sales_age = df.groupby(['Age Group'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)
sns.barplot(x='Age Group',y='Amount',data=sales_age,palette='rainbow',hue='Age Group')
```

<span style="color: blue; font-size: 30px; font-weight: bold;">STATE</span>


```python
sales_state = df.groupby(['State'],as_index=False)['Orders'].sum().sort_values(by='Orders',ascending=False).head(10)

sns.barplot(x='State',y='Orders',data=sales_state,palette='colorblind',hue='State')
#sns.set(rc={'figure.figsize':(15,5)})
plt.figure(figsize=(15,5))

```

```python
sales_state = df.groupby(['State'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False).head(10)

sns.barplot(x='State',y='Amount',data=sales_state,palette='deep',hue='State')
#sns.set(rc={'figure.figsize':(15,5)})
plt.figure(figsize=(15,5))
```

<span style="color: blue; font-size: 30px; font-weight: bold;">MARITAL STATUS</span>


```python
ax = sns.countplot(x='Marital_Status',data=df,palette='Set1',hue='Marital_Status')
plt.figure(figsize=(7,5))

for bars in ax.containers:
    ax.bar_label(bars)
```

```python
sales_state = df.groupby(['Marital_Status','Gender'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)

plt.figure(figsize=(6,5))
sns.barplot(x='Marital_Status',y='Amount',data = sales_state,palette='rainbow',hue='Gender')
```

<span style="color: blue; font-size: 30px; font-weight: bold;">OCCUPATION</span>

```python
ax = sns.countplot(x='Occupation',data=df,palette='muted',hue='Occupation')

plt.figure(figsize=(50,20))
plt.tight_layout()

for bars in ax.containers:
    ax.bar_label(bars)
           
```

```python
sales_state = df.groupby(['Occupation'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)

sns.barplot(x='Occupation',y='Amount',data=sales_state,palette='Set1',hue='Occupation')
plt.figure(figsize=(30,5))
plt.tight_layout()
```

<span style="color: blue; font-size: 30px; font-weight: bold;">PRODUCT CATEGORY</span>

```python
ax = sns.countplot(x='Product_Category',data=df,palette='bright',hue='Product_Category')
plt.figure(figsize=(20,5))
plt.tight_layout

for bars in ax.containers:
           ax.bar_label(bars)
 
```

```python
sales_state = df.groupby(['Product_Category'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False).head(10)

plt.figure(figsize=(20,5))
sns.barplot(data = sales_state, x = 'Product_Category',y= 'Amount',palette='Set1',hue='Product_Category')
```

```python
sales_state = df.groupby(['Product_ID'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False).head(10)

plt.figure(figsize=(20,5))
sns.barplot(data = sales_state, x = 'Product_ID',y= 'Amount',palette='colorblind',hue='Product_ID')
```
